```python
from jamo import h2j, j2hcj
def get_jongsung_TF(sample_text):
    sample_text_list = list(sample_text)
    last_word = sample_text_list[-1]
    last_word_jamo_list = list(j2hcj(h2j(last_word)))
    last_jamo = last_word_jamo_list[-1]

    jongsung_TF = "T"

    if last_jamo in ['ã…', 'ã…‘', 'ã…“', 'ã…•', 'ã…—', 'ã…›', 'ã…œ', 'ã… ', 'ã…¡', 'ã…£', 'ã…˜', 'ã…š', 'ã…™', 'ã…', 'ã…', 'ã…¢', 'ã…,ã…”', 'ã…Ÿ', 'ã…–', 'ã…’']:
        jongsung_TF = "F"

    return jongsung_TF
```

```python
with open("./user-dic/nnp.csv", 'r', encoding='utf-8') as f:
  file_data = f.readlines()

for word in word_list:
  jongsung_TF = get_jongsung_TF(word)

  line = '{},,,,NNP,*,{},{},*,*,*,*,*\n'.format(word, jongsung_TF, word)

  file_data.append(line)
```

```python
# customí•œ ë°©ì‹ìœ¼ë¡œ ë¶ˆìš©ì–´ ì„¤ì •
stopwords = "ë° ë§Œí¼ í•¨ ê·¸ ê°ê° ê´€ë ¨ ë˜í•œ ê²° ê´€ ì²´ ìœ„ ë¶€ ë° ê¸° ì¸µ í•˜ë‚˜ í†µí•´ ê´€í•œ ìœ„í•œ ëŒ€í•œ í•œ ë§ˆë ¨ ì œ ì´ ìˆ í•˜ ìƒ ìš© ê²ƒ ë“¤ ê·¸ ë˜ ìˆ˜ ì´ ë³´ ì•Š ì—† ë‚˜ ì‚¬ëŒ ì£¼ ì•„ë‹ˆ ë“± ê°™ ìš°ë¦¬ ë•Œ ë…„ ê°€ ì§€ ëŒ€í•˜ ì˜¤ ë§ ì¼ ê·¸ë ‡ ìœ„í•˜ ì € ìœ„í•œ ì „ ë‚œ ì¼ ê±¸ ë­ ì¤„ ë§Œ ê±´ ë¶„ ê°œ ë ì¼ ì´ê±° ë²ˆ ì¤‘ ë“¯ ë•Œ ê²Œ ë‚´ ë§ ë‚˜ ìˆ˜ ê±° ì  ê²ƒ ì˜ ê°€ ì´ ì€ ë“¤ ëŠ” ì¢€ ì˜ ê± ê³¼ ë„ ë¥¼ ìœ¼ë¡œ ì ì— ì™€ í•˜ë‹¤ ì„ ì•„ ê·¸ ì¢€"
stopwords = list(set(stopwords.split(" ")))
print(stopwords)
```
